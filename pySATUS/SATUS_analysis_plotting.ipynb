```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import SymLogNorm
```

```python
fname = "test_outputs/TEST_OUTPUT_0005.h5"
ds = h5py.File(fname, "r")
```

```python
x1 = ds["MeshData"]["x1"][:]
x2 = ds["MeshData"]["x2"][:]
```

```python
cmap_dict = {
    "PhotCnt"    : "viridis",
    "AbsCoeff"   : "hot",
    "ScatCoeff"  : "plasma",
    "AnisoCoeff" : "jet"
}
```

```python
fig, ax = plt.subplots(2,2,figsize=(10,10))

im1 = ax[0,0].imshow(ds["MeshData"]["PhotCnt"][:,:], extent=(x1.min(),x1.max(),x2.min(),x2.max()),
                    cmap = cmap_dict["PhotCnt"])

divider1 = make_axes_locatable(ax[0,0])
cax1 = divider1.append_axes("right",size="5%",pad=0.05)
cbar1 = plt.colorbar(im1, cax=cax1)

im2 = ax[0,1].imshow(ds["MeshData"]["AbsCoeff"][:,:], extent=(x1.min(),x1.max(),x2.min(),x2.max()),
                    cmap = cmap_dict["AbsCoeff"])

divider2 = make_axes_locatable(ax[0,1])
cax2 = divider2.append_axes("right",size="5%",pad=0.05)
cbar2 = plt.colorbar(im2, cax=cax2)

im3 = ax[1,0].imshow(ds["MeshData"]["AnisoCoeff"][:,:], extent=(x1.min(),x1.max(),x2.min(),x2.max()),
                    cmap = cmap_dict["AnisoCoeff"])

divider3 = make_axes_locatable(ax[1,0])
cax3 = divider3.append_axes("right",size="5%",pad=0.05)
cbar3 = plt.colorbar(im3, cax=cax3)

im4 = ax[1,1].imshow(ds["MeshData"]["ScatCoeff"][:,:], extent=(x1.min(),x1.max(),x2.min(),x2.max()),
                    cmap = cmap_dict["ScatCoeff"])

divider4 = make_axes_locatable(ax[1,1])
cax4 = divider4.append_axes("right",size="5%",pad=0.05)
cbar4 = plt.colorbar(im4, cax=cax4)

ax[0,0].set_title("Photon Counts")
ax[0,1].set_title("Absorption Coefficient")
ax[1,0].set_title("Anisotropy of Scattering")
ax[1,1].set_title("Scattering Coefficient")
```

# Photon Count plot

```python
fig, ax = plt.subplots(1,1,figsize=(8,8))

im1 = ax.imshow(ds["MeshData"]["PhotCnt"][:,:], extent=(x1.min(),x1.max(),x2.min(),x2.max()),
                    cmap = cmap_dict["PhotCnt"])

divider1 = make_axes_locatable(ax)
cax1 = divider1.append_axes("right",size="5%",pad=0.05)
cbar1 = plt.colorbar(im1, cax=cax1)

ax.set_xlabel("x1 [mm]", fontsize=16)
ax.set_ylabel("x2 [mm]", fontsize=16)
ax.set_title("Photon Counts in Space", fontsize=16)
```

```python
for num in range(100):
    fname = f"test_outputs/TEST_OUTPUT_{str(num).zfill(4)}.h5"
    ds = h5py.File(fname, "r")
    
    x1 = ds["MeshData"]["x1"][:]
    x2 = ds["MeshData"]["x2"][:]
    
    fig, ax = plt.subplots(1,1,figsize=(8,8),facecolor="white")

    im1 = ax.imshow(ds["MeshData"]["PhotCnt"][:,:], extent=(x1.min(),x1.max(),x2.min(),x2.max()),
                        cmap = cmap_dict["PhotCnt"])

    divider1 = make_axes_locatable(ax)
    cax1 = divider1.append_axes("right",size="5%",pad=0.05)
    cbar1 = plt.colorbar(im1, cax=cax1)

    ax.set_xlabel("x1 [mm]", fontsize=16)
    ax.set_ylabel("x2 [mm]", fontsize=16)
    ax.set_title("Photon Counts in Space", fontsize=16)
    
    plt.savefig(f"test_outputs/count_plot_{str(num).zfill(4)}.png")
    plt.close()
```
